```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#binomial distribution

data= np.random.binomial(5, .6, 100)

mean = np.mean(data)
sd = np.std(data)
print('Mean :' + str(mean))
print('SD :' + str(sd))
#Plot a histogram.
plt.hist(data)
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
# Print the histogram
plt.show()

```

```python
# # Generate a gamma distribution with shape =5 and scale = 1
gamma = np.random.gamma(5,.6, 100)

# Plot a histogram.
plt.hist(gamma)
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
# Print the histogram.
plt.show()
```

```python
# exponential
gamma = np.random.exponential(.6, 100)

# Plot a histogram.
plt.hist(gamma)
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
# Print the histogram.
plt.show()
```

```python
# ecommerce poisson
ecommerce = pd.DataFrame()
ecommerce['spending'] = np.random.poisson(75, 100)

plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')

# Plot a histogram.
plt.hist(ecommerce['spending'])
plt.show()
```

```python
#standard_normal
ecommerce = pd.DataFrame()
ecommerce['spending'] = np.random.standard_normal( 100)
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.hist(ecommerce['spending'])
plt.show()
```

```python
#lognormal
lognormala = np.random.lognormal(50,1, 100)
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.hist(lognormala)
plt.show()
```

```python
var1 = np.random.normal(5,.5, 100)
var2 = np.random.normal(10,1, 100)
var3 = var1+var2

mean = np.mean(var3)
sd = np.std(var3)

plt.hist(var3)
plt.axvline(x=mean,color='lightgrey')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
```
